# Problem Statement 
A credit card is one of the most used financial products to make online purchases and payments. Though the Credit cards can be a convenient way to manage your finances, they can also be risky. Credit card fraud is the unauthorized use of someone else's credit card or credit card information to make purchases or withdraw cash.
It is important that credit card companies are able to recognize fraudulent credit card transactions so that customers are not charged for items that they did not purchase. 
The dataset contains transactions made by credit cards in September 2013 by European cardholders. This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.
We have to build a classification model to predict whether a transaction is fraudulent or not.

## 2.4. Outliers Treatment
Handling outliers is a crucial part of data preprocessing, as outliers can significantly affect the performance of some models.

```python

# List of all features except for 'Class'
features = [col for col in df.columns if col != 'Class']

# Define the number of rows and columns for the subplot grid
n_cols = 3
n_rows = int(np.ceil(len(features) / n_cols))

# Create a figure and a set of subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4 * n_rows))
fig.tight_layout(pad=4.0)

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Plot each feature in a separate subplot
for idx, col in enumerate(features):
    sns.boxplot(x='Class', y=col, data=df, ax=axes[idx])
    axes[idx].set_title(f'Boxplot of {col} by Class')
    axes[idx].set_xlabel('Class (0: Non-Fraud, 1: Fraud)')
    axes[idx].set_ylabel(col)

# Remove any empty subplots if features don't fill the grid completely
for i in range(len(features), len(axes)):
    fig.delaxes(axes[i])

# Show the plot
plt.show()
```

### Handling Outliers for non_fraudulent transactions

```python
# Capping all columns having non-fraudulent transactions except 'Class' and 'Time'
for col in df.columns:
    if col not in ['Class','Time']:
        # Calculate the IQR for the feature in non-fraudulent transactions
        q1 = df[df['Class']==0][col].quantile(0.25)
        q3 = df[df['Class']==0][col].quantile(0.75)
        iqr = q3 - q1
        
        # Calculate the upper and lower limits for outlier capping
        upperlimit = q3 + 1.5 * iqr
        lowerlimit = q1 - 1.5 * iqr
        
        # Apply capping to the non-fraudulent transactions
        df.loc[(df['Class'] == 0) & (df[col] > upperlimit), col] = upperlimit
        df.loc[(df['Class'] == 0) & (df[col] < lowerlimit), col] = lowerlimit
```

```python
# checking again if outliers are capped 
features = [col for col in df.columns if col != 'Class']

# Define the number of rows and columns for the subplot grid
n_cols = 3
n_rows = int(np.ceil(len(features) / n_cols))

# Create a figure and a set of subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4 * n_rows))
fig.tight_layout(pad=4.0)

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Plot each feature in a separate subplot
for idx, col in enumerate(features):
    sns.boxplot(x='Class', y=col, data=df, ax=axes[idx])
    axes[idx].set_title(f'Boxplot of {col} by Class')
    axes[idx].set_xlabel('Class (0:Non-Fraud, 1:Fraud)')
    axes[idx].set_ylabel(col)

# Remove any empty subplots if features don't fill the grid completely
for i in range(len(features), len(axes)):
    fig.delaxes(axes[i])

# Show the plot
plt.show()
```

# 3. SCORE CARD

#### A summary of model performance metrics and evaluation results.

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_auc_score
```

```python
# Create an empty DataFrame to store the scores for various algorithms
score_card = pd.DataFrame(columns=['Model', 'AUC Score', 'Precision Score', 'Recall Score',
                                   'Accuracy Score', 'f1-score'])

# Function to append the result table for all performance scores
# Performance measures considered for model comparison are 'AUC Score', 'Precision Score', 'Recall Score', 'Accuracy Score', and 'f1-score'
def update_scorecard(model, test, pred, prob):
    global score_card
    # Create a new dataframe 'new_scorecard'
    new_scorecard = pd.DataFrame({
        'Model': [model],
        'AUC Score': [roc_auc_score(test, prob)],
        'Precision Score': [precision_score(test, pred)],
        'Recall Score': [recall_score(test, pred)],
        'Accuracy Score': [accuracy_score(test, pred)],
        'f1-score': [f1_score(test, pred)]
    })

    # Append the new scorecard to the global score_card DataFrame
    score_card = pd.concat([score_card, new_scorecard], ignore_index=True)

# Function to remove a model's scores from the score_card by index
def remove_model_by_index(index):
    global score_card
    # Drop the specified row by index and reset the index
    score_card = score_card.drop(index).reset_index(drop=True)
```

## 5.1.2 AUC ROC Curve

The AUC ROC curve is used to evaluate the performance of a classification model. The ROC curve plots the True Positive Rate (TPR) against the False Positive Rate (FPR), while the AUC (Area Under the Curve) measures the model's ability to distinguish between classes. A higher AUC indicates better model performance.

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
# Predict probabilities for the test set
y_prob = log_model.predict_proba(X_test)[:, 1]  # Probability of positive class

# Calculate the AUC score
auc_score = roc_auc_score(y_test, y_prob)
print(f'AUC Score: {auc_score:.4f}')

# Compute the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Plot the AUC-ROC curve
plt.figure(figsize=(8, 6))
sns.lineplot(x=fpr, y=tpr, label=f'AUC = {auc_score:.4f}', color='blue')
plt.plot([0, 1], [0, 1], linestyle='--', color='red')  # Dashed diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Logistic Regression Initial Model')
plt.legend(loc='best')
plt.grid(True)  # Adding grid for better visualization
plt.show()
```

### 5.1.3 Hyperparameter Tuning¶

Hyperparameter tuning is the process of optimizing the settings or configurations of a model (such as learning rate, number of trees, or depth of a tree) to improve its performance. Techniques like Grid Search or Random Search are commonly used to find the best combination of hyperparameters.


```python
# Define the model
log_m = LogisticRegression(random_state=42)

# Define the hyperparameters and their values
param_grid = {
    'solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10, 100],
    'max_iter': [100, 200, 300, 500, 1000]
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define LogisticRegression model
log_m = LogisticRegression()

# Define parameter grid for grid search
param_grid = {
    'penalty': ['l1', 'l2'],  # Allow both l1 and l2 penalties
    'C': [0.1, 1, 10],
    'solver': ['liblinear'],  # Use solver that supports both penalties
}

# Perform grid search
grid_search = GridSearchCV(
    estimator=log_m,
    param_grid=param_grid,
    scoring='f1',
    cv=5, 
    n_jobs=-1
)

# Fit the grid search
grid_search.fit(X_train, y_train)

# Get the best parameters and best score
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best F1 Score: {grid_search.best_score_:.4f}")
```

## 5.7 Ada Boost Classifier

AdaBoost (Adaptive Boosting) is an ensemble machine learning algorithm that combines multiple weak learners (typically decision trees) to create a strong classifier. It works by adjusting the weights of incorrectly classified instances, focusing more on difficult cases. AdaBoost iteratively builds models to improve accuracy and reduce errors.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# Initialize the base model (DecisionTreeClassifier as base estimator)
base_model = DecisionTreeClassifier(max_depth=1, random_state=42)

# Initialize the AdaBoost model
ada_model = AdaBoostClassifier(base_estimator=base_model, random_state=42)

# Train the model
results_ada = ada_model.fit(X_train, y_train)

# Make predictions
y_pred_ada = results_ada.predict(X_test)
y_prob_ada = results_ada.predict_proba(X_test)[:, 1]
```

## Business Insights:
### High Precision for Small Transactions:
For banks with smaller average transaction values, precision is critical. Mislabeling genuine transactions as fraudulent can be mitigated by verifying flagged transactions through customer contact. However, low precision increases manual workload, making this process inefficient.

### High Recall for Large Transactions:
For banks handling large transaction values, recall is more important. Missing a fraudulent high-value transaction (low recall) can result in significant financial losses. Therefore, it is essential to maximize recall to catch as many fraudulent transactions as possible.

### Model Recommendation:
After testing various models, the Logistic Regression model, trained on a balanced dataset using the SMOTE technique, performed well. It achieved a high ROC score and recall, making it suitable for detecting fraudulent transactions. Additionally, its simplicity and interpretability make it easy to explain to business stakeholders.

This approach balances the need for precision and recall, depending on transaction size, to minimize financial risk and operational burden.
